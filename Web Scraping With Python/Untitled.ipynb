```python
#importing BeautifulSoup class from bs4 library
from bs4 import BeautifulSoup
#importing requests
import requests
#importing pandas
import pandas as pd
```

```python
#Getting response
response = requests.get("https://en.wikipedia.org/wiki/List_of_bicycle-sharing_systems")
web_data = response.text
print(web_data)
```

```python
#creating soup as an object
soup = BeautifulSoup(web_data, "html.parser")
```

```python
#Finding the table element in the web
table = soup.find(name="table")
print(table)
```

```python
#Extracing only <tr> tags
table_rows = soup.table.find_all(name="tr")
```

```python
#Creating an empty lists to store the data within <tr> tags
table_data_list = []
#Looping over the tr tags and storing datas
for tr in table_rows:
    t_datas = tr.find_all(name="td")
    t_heads = tr.find_all(name="th")
    head = [th.getText() for th in t_heads]
    table_data_list.append(head)
    row = [td.getText().rstrip('\n') for td in t_datas]
    table_data_list.append(row)
print(table_data_list)
```

```python
#converting the list into dataframe with the help of pandas library
df = pd.DataFrame(table_data_list)
print(df)
```
